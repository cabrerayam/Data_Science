```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Cargar este archivo en Python. Realizar estadísticas descriptivas básicas
df_hr = pd.read_csv('/content/HRDataset_v14 (1).csv', encoding='latin-1')
df_hr.head()
```

```python
df_hr.describe()
```

```python
#Realizar un histograma con los salarios. Qué rango(s) de salarios son los más populares?

fig, ax = plt.subplots(figsize=(10, 5))
ax.hist(df_hr['Salary'].values.flatten(), bins=10, edgecolor = 'black', linewidth=1)
ax.set_title('Histograma de salarios')
ax.set_xlabel('Intervalos de salarios')
ax.set_ylabel('Frecuencia absoluta')

#Los rangos de salarios mas populares son 50000 y 100000
```

```python
# Realizar un gráfico de violín con los salarios, discriminados por género.
ax = sns.catplot(data=df_hr, kind='violin', x='GenderID', y='Salary', split=True)
ax.set(xlabel='Generos', ylabel='Salarios', title='Categorización de salarios por genero')
```

```python
#Graficar la serie de tiempo correspondiente a la fecha de contratación (DateofHire)
df_hire = df_hr[['DateofHire', 'EmpID',]].copy()
df_hire= df_hire.drop('DateofHire', axis='columns')
df_hire.head()
```

```python
df_hire = df_hire.groupby('DateofHire').count()
```

```python
x = df_hire.index
y= df_hire.sum(axis = 'columns')

fig, ax = plt.subplots(figsize=(20, 5))
ax.plot(x, y)
ax.set_xlabel ('Fecha de contratacion')
ax.set_ylabel ('Cantidad de empleados')
ax.set_title('Cantidad de Contrataciones por Fecha')
ax.legend()
```
